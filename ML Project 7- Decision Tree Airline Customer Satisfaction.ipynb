## Part-2: Working with models

```python
#1) Perform encoding in columns Gender, Customer_Type, Type_of_Travel, and Class.
```

```python
data["Gender"].unique() #label 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
data["Gender"]=encoder.fit_transform(data["Gender"])
```

```python
data["Customer_Type"].unique()#label
```

```python
data["Customer_Type"]=encoder.fit_transform(data["Customer_Type"])
```

```python
data["Type_of_Travel"].unique() # one hot 
```

```python
data=pd.get_dummies(data,columns=["Type_of_Travel"])
```

```python
data["Class"].unique() #one hot
```

```python
data=pd.get_dummies(data,columns=["Class"])
```

```python
data
```

```python
data.isnull().sum()
```

```python
#3) Create the features and target Data
```

```python
x=data.drop("satisfaction",axis=1)
y=data["satisfaction"]
```

```python
data
```

```python
#4) Perform scaling on features data
#5) Split the data in training and testing sets
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#scaling 
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DC=DecisionTreeClassifier()
```

```python
DC.fit(x_train,y_train)
```

```python
DC.score(x_train,y_train)*100
```

```python
DC.score(x_test,y_test)*100
```

```python
yp=DC.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,accuracy_score,f1_score,classification_report
```

```python
confusion_matrix(yp,y_test)
```

```python
accuracy_score(yp,y_test)*100
```

```python
precision_score(yp,y_test)
```

```python
recall_score(yp,y_test)
```

```python
f1_score(yp,y_test)
```

```python
print("Classification Report:",classification_report(yp,y_test))
```
